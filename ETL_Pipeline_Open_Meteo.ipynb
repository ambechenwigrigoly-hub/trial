ETL Pipeline : weather

This notebook shows a ETL workflow using:

.requests to extract data from a public API
.json() to convert the API response into Python data
pandas to transform the data into a table
.to_csv() to load / save the result
.matplotlib to create simple charts

Using the public open metro API. No password or API key is needed

1. Impot of libraries: request,json,pandas and matplotlib

```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

2.EXTRACT: Request data from the API

Collect metro data about the weather

```python
url = "https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&current=temperature_2m,wind_speed_10m&hourly=temperature_2m,relative_humidity_2m,wind_speed_10m&forecast_days=1&timezone=auto"

Parameters= {
  "current": {
    "time": "2022-01-01T15:00",
    "temperature_2m": 2.4,
    "wind_speed_10m": 11.9,
  },
  "hourly": {
    "time": ["2022-07-01T00:00","2022-07-01T01:00"],
    "wind_speed_10m": [3.16,3.02,3.3,3.14,3.2,2.95],
    "temperature_2m": [13.7,13.3,12.8,12.3,11.8],
    "relative_humidity_2m": [82,83,86,85,88,88,84,76]
  }
}

response = requests.get(url)

print(response.status_code)
```

3.Convert response to JSON data

Converting the API response into a python object i.e data_json

```python
data_json = response.json()

print(data_json)
```

4.Create a pandas data frame

Converting the data_json to a pivot table:pandas data frame

```python
import pandas as pd

data_json = response.json()
df = pd.DataFrame(data_json['hourly'])

df.head()
```

5.TRANSFORM:Rename columns

```python
df = df.rename(columns={
    "time": "ToD",
    "temperature_2m": "Temp_2m",
    "relative_humidity_2m": "RH_2m",
    "wind_speed_10m": "WS_10m"
})

df.head()
```

6.LOAD : Save the final pandas data table as CSV file

It is the end step of the ETL

```python
df.to_csv("open_metro_data.csv", index=False)

print("CSV file saved: open_metro_data.csv")
```
